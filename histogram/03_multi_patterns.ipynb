```python
import matplotlib.pyplot as plt
import numpy as np

import os
os.chdir("/home/wtian/play_ground/plot/histogram")

# 基对变化模式和相应的计数
mutations = ['CG>GC', 'GC>CG', 'CG>AC', 'AC>CG', 'AT>CG', 'GT>AC', 'CG>TA', 'CG>AT', 'GG>AA', 'CA>TG']
transversion = np.array([2000, 1800, 1600, 1400, 1000, 800, 600, 400, 200, 180000])
transition = np.array([4000, 3600, 3200, 2800, 2000, 1600, 1200, 800, 400, 360000])
transversion_transition = np.array([1000, 900, 800, 700, 500, 400, 300, 200, 100, 90000])

# 创建图表
fig, ax = plt.subplots()

# 添加柱状图
index = np.arange(len(mutations))
bar_width = 0.2

rects1 = ax.bar(index, transversion, bar_width, label='Transversion')
rects2 = ax.bar(index + bar_width, transition, bar_width, label='Transition')
rects3 = ax.bar(index + 2*bar_width, transversion_transition, bar_width, label='Transversion+Transition')

ax.set_xlabel('Mutation Pattern')
ax.set_ylabel('Number of MNV')
ax.set_title('Number of MNV by mutation type and pattern')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(mutations, rotation=45, ha="right")
ax.set_yscale('log')  # 使用对数刻度
ax.legend()

# 显示图表
# plt.show()

plt.savefig('03_mnv_bar_chart.png', dpi=300, bbox_inches='tight')
```
